```python
# Purely integer-location based indexing for selection by position.
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iloc.html

X=final_dataset.iloc[:,1:]
y=final_dataset.iloc[:,0]

print(X.head())
print("-------------------------------------------------------\n")
print(y.head())
```

```python
X['Owner'].unique()
```

```python
from sklearn.ensemble import ExtraTreesRegressor

# 极其随机的森林回归 ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X,y)

# The importances of each feature
print(model.feature_importances_)
```

```python
######## Plot graph of feature importances

# Creating a Series by passing a list of values
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
print(feat_importances)
print("\n")

# Return the largest n elements.
feat_importances_largest5 = feat_importances.nlargest(5)
print(feat_importances_largest5)

# Make plots of Series or DataFrame.
# https://pandas.pydata.org/docs/reference/api/pandas.Series.plot.html
# kind='barh’ : horizontal bar plot
import matplotlib.pyplot as plt
feat_importances_largest5.plot(kind='barh')
plt.show()
```

```python
########################################## 超参数搜索开始

from sklearn.model_selection import train_test_split

# Split arrays or matrices into random train and test subsets
# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# print(X_train.head())
# print(y_train.head())
```

```python
import numpy as np

# numpy.linspace
# Return evenly spaced numbers over a specified interval.
# https://numpy.org/devdocs/reference/generated/numpy.linspace.html?highlight=linspace#numpy.linspace

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]

# Number of features to consider at every split
max_features = ['auto', 'sqrt']

# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]

# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]

# 定义超参数的分布（或取值范围，或搜索空间）
param_dist = {'n_estimators': n_estimators,
              'max_features': max_features,
              'max_depth': max_depth,
              'min_samples_split': min_samples_split,
              'min_samples_leaf': min_samples_leaf}
print(param_dist)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Use the random grid to search for best hyperparameters
# First create the base model to tune
regressor = RandomForestRegressor()
```

```python
from sklearn.model_selection import RandomizedSearchCV

# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations
random_search = RandomizedSearchCV(estimator = regressor, 
                                   param_distributions = param_dist,
                                   scoring = 'neg_mean_squared_error', 
                                   n_iter = 10, # 从param_distributions中sample出多少个参数集合
                                   cv = 5, 
                                   verbose = 2, 
                                   random_state = 42,
                                   n_jobs = 1) # 并行处理的任务数
```

```python
random_search.fit(X_train,y_train) # 搜索最佳超参数
```

```python
print(random_search.best_params_)
print(random_search.best_score_)
print(random_search.best_estimator_)

########################################## 超参数搜索结束
```

```python
predictions = random_search.predict(X_test)
```

```python
import seaborn as sns

# Figure-level function for drawing distribution plots
sns.displot(y_test-predictions)
```

```python
import seaborn as sns

# Axes-level function for plotting histograms to show distributions of datasets.
sns.histplot(y_test-predictions)
```

```python
import matplotlib.pyplot as plt

# A scatter plot of y vs. x
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html
plt.scatter(y_test,predictions)
plt.xlabel("real value")
plt.ylabel("prediction value");
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
